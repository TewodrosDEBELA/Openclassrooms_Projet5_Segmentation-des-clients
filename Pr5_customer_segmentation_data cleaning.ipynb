#### Fusioner entre les donnée Translation et Products

```python
# On utilise les listes des produits en anglaise et on evite les colonnées qui sont pas pertinents.
# Renomer les nom de produits par les catégorie. 

products_merged=pd.merge(translation,products, on="product_category_name", how="left").drop(['product_name_lenght','product_category_name','product_description_lenght','product_photos_qty','product_weight_g','product_length_cm','product_height_cm','product_width_cm'],axis=1)

products_merged['product_category_name_english'] =products_merged[
        'product_category_name_english'].map(dict_categories)
products_merged
```

#### Fusioner entre les trois df afin d'avoir une df de base

```python

df_merge = pd.merge(customers, 
                           pd.merge(order_items,
                                    pd.merge(order_payments,
                                             orders),on="order_id"), on='customer_id')
df_merge
```

### score commentaire par chaque clients


```python
df_merge=pd.merge(df_merge,order_reviews)
Review=df_merge.groupby("customer_unique_id")["review_score"].mean()
Review=Review.rename('mean_review_score', inplace=True).to_frame()

Review
```

### numero de commande 

```python
Numero_de_commande = df_merge.groupby('customer_unique_id')['order_id'].count()
Numero_de_commande=Numero_de_commande.rename('total_orders', inplace=True).to_frame()
    
Numero_de_commande
```

### catégorie la plus achetée

```python
Categorie = pd.merge(df_merge, products_merged).sort_values(
        ['customer_unique_id','product_category_name_english'], ascending=False).groupby(
        ['customer_unique_id','product_category_name_english']).head(1)[[
        'customer_unique_id', 'product_category_name_english']]
Categorie.columns = ['customer_unique_id', 'most_purchased_item']
Categorie.set_index('customer_unique_id', inplace=True)
```

```python
Categorie
```

### details de la categorie

```python

Details_categorie = pd.merge(df_merge, products_merged).groupby(["customer_unique_id", "product_category_name_english"])\
                        .agg({"order_item_id": "count"}).unstack()

Details_categorie.columns =  Details_categorie.columns.droplevel(0)

# Ils existent les Nan comme on decompose les catagorie par chaque client

Details_categorie.fillna(0, inplace=True)


Details_categorie.reset_index(inplace=True)

print( Details_categorie.shape)

Details_categorie.head(10)
```

### montant moyen des achats

```python
Achats_Moyenne =df_merge.groupby(
        ['customer_unique_id', 'order_id'])['price'].sum().groupby(['customer_unique_id']).mean()
Achats_Moyenne=Achats_Moyenne.rename('mean_purchases', inplace=True).to_frame()
Achats_Moyenne
```

### delai livraison moyenne

```python
df_merge['mean_delivery_days'] = df_merge['order_delivered_customer_date'] - df_merge[
        'order_purchase_timestamp']

# afin d'avoir une valuer positive, on ajoute+1 au cas ou la difference est negative

df_merge['mean_delivery_days']  =df_merge['mean_delivery_days'].dt.days
Delai_delivery_Moyenne=df_merge.groupby('customer_unique_id')['mean_delivery_days'].mean().to_frame()
Delai_delivery_Moyenne
```

```python
df_merge=df_merge[['customer_unique_id','customer_city','order_status','payment_sequential','payment_type','payment_installments']].drop_duplicates(subset=['customer_unique_id'], keep='first')
df_merge=pd.merge(df_merge,Details_categorie,how="left", on="customer_unique_id")
df_merge
```

```python
df_merge1=pd.merge(Numero_de_commande,df_merge,how="inner", on="customer_unique_id")
df_merge2=pd.merge( df_merge1,Review,how="inner", on="customer_unique_id")                          
df_merge3=pd.merge(df_merge2, Categorie,how="inner", on="customer_unique_id") 
df_merge4=pd.merge(df_merge3, Achats_Moyenne,how="inner", on="customer_unique_id")
df_merge_last=pd.merge(df_merge4, Delai_delivery_Moyenne,how="inner", on="customer_unique_id").drop_duplicates(subset=['customer_unique_id'], keep='first')
```
