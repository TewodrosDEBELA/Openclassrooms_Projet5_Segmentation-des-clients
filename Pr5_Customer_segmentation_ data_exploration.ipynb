```python
import pandas as pd
import numpy as np
from numpy import NaN
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sn
```

```python
df=pd.read_csv("df_client_segmentation.csv")
df
```

```python
df['order_purchase_timestamp']=df['order_purchase_timestamp'].astype('datetime64[ns]')
```

```python

%matplotlib inline
for column in df.select_dtypes(['int32', 'float64']).columns:
    f, axes = plt.subplots(1,2, figsize=(12,4))
    titre = 'Distribution de ' + str(column)
    plt.title(titre)
    sns.distplot(df[column], bins=30, ax=axes[0])
    titre = 'Distribution de ' + str(column)
    plt.title(titre)
    sns.boxplot(df[column], ax=axes[1])
    plt.show()
```

### most purchased item

```python

df_hist1=df["most_purchased_item"].value_counts().nlargest(10).sort_values(ascending=True)
df_hist1.plot(kind="barh",figsize=(15,5), fontsize=14, color='royalblue')
plt.title("Top 10 most purchased item",fontsize=14)
plt.xlabel('Count',fontsize=14)
plt.ylabel('most_purchased_item',fontsize=14)
```

### number of orders by months

```python
# Max..Mois de mai et Aoute.
# Min..Mois Septembre, Octobre et décembre
# Il est un peu inattendu que le nombre de commandes diminue vers la fin de l'année.

df.groupby(df['order_purchase_timestamp'].dt.month)\
    .agg({"customer_unique_id": "nunique"})\
    .plot(figsize=(12,8), kind="bar",
          title="Distribution of orders by month",
          ylabel="total_orders",
          xlabel="Months",
          legend=False)
plt.xticks(np.arange(0,12), ['Jan','Feb','Mar','Apr','May','Jun',
                             'Jul','Aug','Sept','Oct','Nov','Dec'], 
           rotation='horizontal')
plt.show()
```

### number of orders by days

```python
# Le nombre de commandes diminue à partir de lundi.
# Le samedi affiche le plus petit nombre de commandes tandis que le lundi, nous observons le plus grand nombre de commandes. 
# Cela semble contre-intuitif car on pourrait s'attendre à un nombre élevé de commandes pendant le week-end que les jours de semaine

df.groupby(df['order_purchase_timestamp'].dt.dayofweek)\
    .agg({"customer_unique_id": "nunique"})\
    .plot(figsize=(12,8), kind="bar",
          title="Distribution of orders during the week",
          ylabel="total_orders",
          xlabel="Day of week",
          legend=False)
plt.xticks(np.arange(0,7), ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'], rotation='horizontal')
plt.show()
```

###  linear relations between variables

```python
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(15,7))
ax = sns.heatmap(corr, annot=True,
                 vmin=-1, vmax=1,
                 fmt=".2f", annot_kws={'size':8}, 
                 mask=mask, 
                 center=0, 
                 cmap="coolwarm")
plt.title(f"heatmap of linear relationship\n", 
          fontdict=None)
plt.show()
```
